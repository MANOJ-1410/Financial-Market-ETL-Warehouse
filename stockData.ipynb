```python
close_prices = df['Close']
```

```python
# create a list to store processed data for each stock
processed_data = []
```

```python
for stocks in close_prices.columns:
    temp_df = close_prices[[stocks]].copy()
    temp_df.columns = ['close_price']
    temp_df['stocks'] = stocks

    #calculate 7-day Moving average
    temp_df['moving_avg_7d'] = temp_df['close_price'].rolling(window=7).mean()

    #calculate daily voltaility_7d (Standard Deviation)
    temp_df['daily_return'] = temp_df['close_price'].pct_change()
    temp_df['volatility_7d'] = temp_df['daily_return'].rolling(window=7).std()

    processed_data.append(temp_df)

final_df = pd.concat(processed_data).dropna()

print(final_df.head(10))
```

```python
def fetch_and_process_stocks(ticker_list, days_back=100):
    # ETL function to fetch & transform stock data.
    end = dt.datetime.now()
    start = end - dt.timedelta(days=days_back)

    # Extraction with error handling
    try:
        raw_data = yf.download(ticker_list,start=start,end=end)
        if raw_data.empty:
            print("Warning: No data found for these tickers.")
            return None
    except Exception as e:
        print(f"Error fetching data: {e}")
        return None
    
    close_prices = raw_data['Close']
    processed_data = []

    # Transformation loop
    for ticker in close_prices.columns:
        # data isolation
        temp_df = close_prices[[ticker]].copy()
        temp_df.columns = ['close_price']

        # adding identifiers
        temp_df['ticker'] = ticker

        # Engineering Derived Features
        temp_df['moving_avg_7d'] = temp_df['close_price'].rolling(window=7).mean()
        temp_df['daily_return'] = temp_df['close_price'].pct_change()
        temp_df['volatility_7d'] = temp_df['daily_return'].rolling(window=7).std()

        temp_df['extracted_at'] = dt.datetime.now()

        processed_data.append(temp_df)

        # consolidation
    final_df = pd.concat(processed_data).dropna()
    return final_df
    
    # execution
my_tickers = ['CBA.AX','NAB.AX','WBC.AX','ANZ.AX']
final_df = fetch_and_process_stocks(my_tickers, days_back=150)

if final_df is not None:
    print("ETL Step 1 Successful!")
    print(final_df.head(10))
```

```python
# setting up a primary key
from sqlalchemy import create_engine , text

connection_string = 'postgresql+psycopg2://myuser:mypassword@localhost:5432/market_data'
engine = create_engine(connection_string)

# creating a table
with engine.connect() as conn:
    conn.execute(text("""
        create table if not exists stock_prices (
            date timestamp,
            ticker text,
            close_price decimal,
            moving_avg_7d decimal,
            daily_return decimal,
            volatility_7d decimal,
            extracted_at timestamp,
            primary key (date, ticker));
            
            create index if not exists idx_ticker_data on stock_prices (ticker, date)
            """))
    conn.commit()

# Upsert (the ETL Logic)
def upsert_to_postgresql(df,table_name,engine):
    # push data to temp staging table
    df.to_sql('temp_staging', engine, if_exists='replace',index=True)

    # execute Upsert
    upsert_query = text(f"""
        insert into {table_name} (date, ticker, close_price, moving_avg_7d, daily_return, volatility_7d, extracted_at)
        select "Date", ticker, close_price, moving_avg_7d , daily_return, volatility_7d, extracted_at
        from temp_staging
        on conflict (date, ticker)
        do update set
            close_price = EXCLUDED.close_price,
            moving_avg_7d = EXCLUDED.moving_avg_7d,
            daily_return = EXCLUDED.daily_return,
            volatility_7d = EXCLUDED.volatility_7d,
            extracted_at = EXCLUDED.extracted_at;
        """)
    
    with engine.connect() as conn:
        conn.execute(upsert_query)
        # clean up staging
        conn.execute(text("drop table temp_staging"))
        conn.commit()

# run
upsert_to_postgresql(final_df,'stock_prices',engine)
print("UPSERT Successfully Done")
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    conn.execute(text("""
        create or replace view stock_analystic_view as 
            select date, ticker, close_price,
            avg(close_price) over (partition by ticker order by date rows between 6 preceding and current row) as sql_moving_avg_7d,
            extracted_at
            from stock_prices;      
    """))
    conn.commit()
```
